## Data analysis

Let's check some statistics about bird's voice in recordings.

```python
import os

bird_voice_lengths = []
bird_voice_in_different_sec = 0

recording_bird = {}

with open(os.path.join('train', 'labels.txt'), 'r') as f:
    for i, line in enumerate(f):
        if i != 0:
            recording_name = line.split(',')[0]
            if recording_name in recording_bird:
                recording_bird[recording_name] = recording_bird[recording_name] + 1
            else:
                recording_bird[recording_name] = 1
            start = float(line.split(',')[1])
            stop = float(line.split(',')[2])
            bird_voice_lengths.append(stop - start)
            if int(start) != int(stop):
                bird_voice_in_different_sec += 1
```

#### Does somebody mark incorrect start/stop time point?
By incorrect I mean start after stop point.

```python
import numpy as np

bird_voice_lengths = np.array(bird_voice_lengths)

np.any(bird_voice_lengths <= 0)
```

Values are correctly marked.

#### Bird's voice length

```python
print(f"Max length: {bird_voice_lengths.max()}")
print(f"Min length: {bird_voice_lengths.min()}")
print(f"Median length: {np.median(bird_voice_lengths)}")
print(f"Mean length: {np.mean(bird_voice_lengths)}")
```

Based on above values, interval length for spectogram can be taken not only for 1s but also for other values e.g. lower than maximum length but higher than mean. I decided to choose 0.1s (~twice mean) 

#### How many lengths are non unique?
Below I will display all non unique values with number of occurences.

```python
import collections

c = collections.Counter(bird_voice_lengths)
MORE_THAN_ONE_OCCURENCE_NUM = sum(np.fromiter(c.values(), dtype=float) > 1)
print("Non unique values:")
display(c.most_common(MORE_THAN_ONE_OCCURENCE_NUM))
```

#### Check how many non-unique values (number of occurence also took into consideration) are higher/lower than median

```python
higher_than_median = 0
higher_than_median_counter = 0
all_occurences = 0
all_occurences_counter = 0

for v in c.most_common(MORE_THAN_ONE_OCCURENCE_NUM):
    if v[0] > np.median(bird_voice_lengths):
        higher_than_median += 1 * v[1]
        higher_than_median_counter += 1
    all_occurences += 1 * v[1]
    all_occurences_counter += 1

print(f"Higher than median: {higher_than_median}. Unique values: {higher_than_median_counter}")
print(f"Lower than median: {all_occurences - higher_than_median}. Unique values: {all_occurences_counter}")
```
